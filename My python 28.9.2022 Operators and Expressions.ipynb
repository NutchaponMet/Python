**Arithmetic Operators**

**Addition, Substraction, Multiplication, Division**

```python
a = 20
b = 5
print(a+b) # addition
print(a-b) # Subtration
print(a*b) # Multiplication
print(a/b) # Division
```

Unary Minus

A unary minus *inverses the polarity* of the numeric operand

```python
a = 2
print(-a)
b = -5
print(-b)
```

### Mudulus

A modulus(`a % b` reades as "modulo b") finds the *remainder* resulting from dividing the left operand (a) with the right operand (b).

```python
x = 25
print(x%4) # since 25 = 4*6+1
```

***We cannot *divide or modulo* with 0***

```python
print(2/0)
```

```python
print(2%0)
```

### Exponentiation

An exponentiation (`a ** b`) finds the value of the left operand (a) *raised to the power of* the value of the right operand (b).

```python
print(2**3)
print(2**6)
print(99**0) #sine any numpbers raised to the power of 0 is 1
print(2**-1)
```

### Floor Division
A floor division (`//`) is a *division* that *round* the result *down* to the *nearest integer*(the next smaller integer). (เป็นการหารแบบปัดเศษให้เป็นจำนวนเต็ม)

```python
print(32 / 3)
print(32 // 3)
print(-32 / 3)
print(-32 // 3)
```

### Comparison Operators
A comparison opertor *compares two values*, it return `True` or `False` 
**ในกรณีนี้จะเป็นการเปลี่ยนเทียบค่าของตัวแปรด้วยจะแสดงผลออกมาเป็นค่าของ `bool`**

```python
a = 20
print(a == 20) # *is equal to ?*
print(a != 20) # is NOT equal to ?
print(a > 20) # is greater than ?
print(a < 20) # is less than ?
print(a >= 20) # is greater than or equal to ?
print(a <= 20) # is less than or equal to ?
```

### Logical Operators
#### จะดำเนินการทางตรรกศาสตร์กับตัวแปรประเภท `bool`
A logical opertor performs a *boolean operation* (AND, OR, NOT).

The logical `and` return `True` if and only if both operands are `True`

```python
p = True
q = False
print(p and p)
print(p and q)
print(q and q)
# มีตัวใดตัวหนึ่งเป็น False ก็จะทำให้ค่าเป็น False
```

The logical `or` operator return `False` if and only if both operands are `False`.

```python
p = True
q = False
print(p or p)
print(p or q)
print(q or q)
# มีตัวใดตัวหนึ่งเป็น True ก็จะทำให้ค่าเป็น True
```

The logical `not` หรือ นิเสธ operator *negates* the boolean value of the operand. 
คือการกลับ T เป็น F กลับ F เป็น T

```python
p = True
q = False
print(not p)
print(not q)
```

โดยปกติจะใช้ *comparison operator* ควบคู่กับ *logical operator*

Usually, *comparison operator* are used together with *logical operator*.

```python
username = input("Tell me your name: ")
print(username == "Nut" or username == "Yo")
```
