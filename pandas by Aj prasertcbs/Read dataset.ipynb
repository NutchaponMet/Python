## read All sheet

```python
url = 'https://github.com/prasertcbs/basic-dataset/blob/master/PisaScore2015.xlsx?raw=true'
df = pd.read_excel(url, sheet_name=None)
df
```

```python
df.keys()
```

### merge sheet

```python
pisa = pd.merge(df['math'], df['sci'], on='Country')
pisa.head(10)
```

```python
pisa = pd.merge(df['math'][['Country','Maths']], df['sci'][['Country','Science']], on='Country')
pisa.head(10)
```

```python
pisa =pd.merge(pisa, df['reading'][['Country','Reading']], on ='Country')
pisa.head(10)
```

```python
pisa['Average'] = (pisa.Maths + pisa.Science + pisa.Reading) / 3
pisa.head(10).round(2)
```

```python
pisa['Rank'] = pisa['Average'].rank(ascending=False, method='min').astype(int)
pisa.sort_values('Rank').head(10).round(2)
```
