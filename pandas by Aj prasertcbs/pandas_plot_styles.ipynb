```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
df = pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/mpg.csv')
```

```python
df.sample(10)
```

```python
df.drv.value_counts()
```

```python
df.drv.value_counts().plot.barh()
```

```python
plt.style.use('ggplot')
df.drv.value_counts().plot.barh()
```

```python
plt.style.available
```

```python
plt.style.use('tableau-colorblind10')
df.drv.value_counts().plot.barh()
```

```python
plt.style.library
```

```python
for s in plt.style.available:
    plt.style.use('default')
    plt.style.use(s)
    df.drv.value_counts().plot.barh(title=s, figsize=(4.5,2))
    plt.show()
```

```python
for s in sorted(plt.style.available):
    plt.style.use('default')
    plt.style.use(s)
    df.drv.value_counts().plot.barh(title=s, figsize=(4.5,2))
    plt.show()
```
