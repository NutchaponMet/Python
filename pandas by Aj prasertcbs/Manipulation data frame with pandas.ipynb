## pandas query rows

```python
import pandas as pd
```

```python
url = 'https://github.com/prasertcbs/tutorial/raw/master/mpg.csv'
df = pd.read_csv(url)
df.sample(10)
```

```python
df.manufacturer == 'audi'
```

```python
df[df.manufacturer == 'audi']
```

```python
df[df.manufacturer != 'audi']
```

```python
df[df.cty > 30]
```

```python
df[df.cty <= 15]
```

```python
df[df.cty >= 15]
```

```python
df[(df.manufacturer == 'toyota') & (df.cty >25)]
```

```python
df[(df.manufacturer == 'toyota') | (df.cty >25)]
```

```python
df[(df.manufacturer == 'toyota') | (df.manufacturer == 'nissan') | (df.manufacturer == 'honda')]
```

```python
df[df.manufacturer.isin(['toyota','nissan','honda'])]
```

```python
df[~df.manufacturer.isin(['toyota','nissan','honda'])]
```

```python
df[df.manufacturer.str.contains('toyota|nissan|honda')].head(5)
```

```python
df[df.manufacturer.isin(['toyota','nissan','Honda'])] # Sensitive case
```

```python
df[df.manufacturer.str.contains('toyota|nissan|Honda', case=False)] # I can compare case insensitive
```
