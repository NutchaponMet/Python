```python
import pandas as pd
```

```python
url = 'https://github.com/prasertcbs/tutorial/raw/master/mpg.csv'
df = pd.read_csv(url)
df.sample(10)
```

```python
df.manufacturer == 'audi'
```

```python
df[df.manufacturer == 'audi']
```

```python
df[df.manufacturer != 'audi']
```

```python
df[df.cty > 30]
```

```python
df[df.cty < 15]
```

```python
df[(df.manufacturer == 'toyota') & (df.cty > 25)]
```

```python
df[(df.manufacturer == 'toyota') | (df.cty > 25)]
```

```python
df[(df.manufacturer == 'toyata') | (df.manufacturer == 'nissan') | (df.manufacturer == 'honda')]
```

```python
# วิธีเขียนแบบย่อ

df[df.manufacturer.isin(['toyata','nissan','honda'])]

# Side-effect ไม่สามารถระบุ case insensitive ได้
```

```python
# วิธีเขียนแบบย่อ
# กรณีที่ทุกยี่ห้อยกเว้นสามรุ่นที่ระบุ

df[~df.manufacturer.isin(['toyata','nissan','honda'])]
```

```python
# กรณีนี้เราสามารถกำหนด case insensitive ได้
df[df.manufacturer.str.contains('toyota|nissan|Honda')].head(10)
```

```python
# กรณีนี้เราสามารถกำหนด case insensitive ได้
df[df.manufacturer.str.contains('toyota|nissan|Honda', case=False)].head(10)
```
