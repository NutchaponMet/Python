### Creating Set

```python
# Create a set using curly brackets.
# ในกรณีนี้มี item ซ้ำ item ที่ซ้ำจะถูกเก็บเป็น item เดียว

sport = {'rugby', 'soccer', 'basketball', 'volleyball', 'softball', 'rugby'}
```

```python
print(type(sport))
print(sport)
```

```python
len(sport)
```

```python
s = {}
```

```python
print(type(s))
print(s)

# Default '{ }' = dict
```

- if you want to create empty set you can use `set()` function.
                s = set(<iterable_object>)

```python
# Example

s = set()

print(s)
print(type(s))
```

```python
# How to resolve by take '[ ]' into 'set()'

#e.g.
s = set([1])

print(s)
```

- iterable odject can be `str()`, `list()` and `tuble()`.

```python
s = set('abc')
print(s)

# เนื่องด้วย Set ไม่มีลำดับ ฉะนั้นผลที่ได้อาจจะไม่เรียงตามลำดับตอนเพิ่ม elements ลงไปใน Set
```

```python
dad_sport = set(['soccer','baseball','basketball','tennis'])
mom_sport = set(['basketball','volleyball','tennis','rowing'])
kid_sport = set(['soccer','tennis','swimming'])
```

- Set **union** returns a new set with distinct items from all the given sets.

1. Using `|` operator.
2. Using `union()` method.

```python
# Using | opertor to union 2 set operands.
dad_sport | mom_sport
```

```python
# Using union() method with iterable object(s) as the argument.

dad_sport.union(mom_sport)
```

```python
dad_sport.union(mom_sport,kid_sport)
```

```python
dad_sport & mom_sport
```

```python
dad_sport&mom_sport&kid_sport
```

```python
dad_sport.intersection(mom_sport,kid_sport)
```

- Set **difference** returns a new set with all items in the given set but not in the other sets.
1. Using `-` operator.
2. Using `difference()` method.

```python
dad_sport - mom_sport
```

```python
dad_sport - mom_sport -kid_sport
```

- Set **symmetric difference** returns a new set of all items in either in A or B, but not both.
1. Using `^` operator.
2. Using `symmetric_difference()` method.

```python
dad_sport^mom_sport
```

```python
dad_sport.symmetric_difference(kid_sport)

# In case variable have not same value 'symmetric_difference' will return 'empty set'
```

### Adding, updating, and deleting items in set.
- Sets are **mutable** collection of **distinct immutable** values that are unordered.

- `add(<immutable object>)` method adds a single objects to the  set.

```python
a = set([1,2,3])
b = {3,4,5}
a.add(4)
```

```python
print(a)
```

- `update(<iterable object>)` method updates the set, adding items from other iterables.

```python
a.update(b)
print(a)
```

```python
a.update([5,6,7], 'xyz')
```

```python
print(a)
```

```python
a | b
```

```python
a = {1,2,3}
b ={4,5,6}
```

```python
a|b
```

```python
a = a|b
```

```python
print(a)
```

```python
a|=b
print(a)
```

- `remove(<item>)` method removes an item from the set. `Error` is raised if the item is not in the set.

```python
a.remove(5)
print(a)
```

- `discard(<item>)` method removes from the set. No exception throws if the item is not in the set.

```python
a.discard(4)
print(a)
```

- `pop()` method deletes and returns a random item from the set. if the set is empty, `Error` is raised. 

```python
a.pop()
```

- `clear()` method deletes all items from the set

```python
print(b)
```

```python
b.clear()
```

```python
print(b)
```

## Iterating through Set
- We can use `for` loop to iterate through sets.

**เราสามารถใช้การวนลูปผ่าน Set ได้**

```python
print(dad_sport)
```

```python
for i in dad_sport:
    print(i)
```

```python
for i in enumerate(dad_sport):
    print(i)
```
