## Tuple
Tuple stores multiple **items** in a single variable, similar to lists. The difference is that tuples are **immutable.

### Creating Tuples
- Tuples can be created using round brackets (parentheses) `( )`

```python
#Create a tuple using a parenthesis.

position = (100, 50)
type(position)
```

```python
len(position)
```

- Tuples can be create using `tuple()` function.

```python
# The tuple() function creates a tuple object.
# Paremeter = iterable object (optional).

t = tuple([1,2,3])
```

```python
print(t)
print(type(t))
```

```python
t = tuple('abc')
print(t)
```

```python
t = tuple((1,2,3))
print(t)
```

- Can tuple be *empty*?

```python
t1 = ()
t2 = tuple()
print(t1, type(t1))
print(t2, type(t2))
```

- Create a tuple with *one* item.

```python
t1 = (1)
print(t1, type(t1))
```

```python
# ถ้าต้องการสร้าง tuple 1 ค่าจะต้องใส่ comma ','
t1 = (1,)
print(t1, type(t1))
```

```python
t2 = tuple(1)
```

```python
t2 = tuple([1])
print(t2, type(t2))
```

### items in a tuple can be *same* or *different* data types. (Heterogenous)

```python
days = ('Sun','Mon','Tue','Wed','Thu','Fri','Sat')

book_info = ('python basics', 39.99, 2021, ('d.amos','d.bader','j.jablonski','f.heisler'))
print(book_info)
print(type(book_info))
```

### Tuple items are "ordered", addressing by an *index*.



Ref:https://pythontutor.com/render.html#mode=display
- To access an item in the tuple using **index**; we use a square bracket: `[ ]` 

```python
print(book_info[0])
print(book_info[-1])
```

### Slicing
- We can also use **slicing** to access a *range of items* by defining the `start`,`stop` an `step` parameters.


```python
print(book_info[:5])
```

### Unpacking tuples
- We can extract the values of tuple's items assigning to variables.

```python
title, price, year = book_info[:-1]
print(title)
print(price)
print(year)
```

```python
position = (100, 50)
x, y = position
```

### Tuple are "immutable"; can not be modified.
But can be reassigned.

```python
print(position)
```

```python
# Tuple can be reassigned
position = (120, 50)
print(position)
```

```python
# Combining tuples using operator '+'
pos = (1,2,3)
position += pos
print(position)
```

```python
# id() function returns the identity of the object.
# Parameter = an object

# Sample
t1 = (1,2,4,3,5)
print(id(t1), t1)
```

```python
t2 = (40,50)
print(id(t2),t2)
```

```python
t1 += t2
print(id(t1), t1)
```

### Tuple are "iterable" objects same List.
- We can **iterate** through tuples.

```python
days = ('Sun','Mon','Tue','Wed','Thu','Fri','Sat')
```

```python
# use for loop
for i in days:
    print(i)
```

```python
# use while() loop
i = 0
while i < len(days):
    print(days[i])
    i += 1
```

```python
# loop through the index number
for i in range(len(days)):
    print(days[i])
```

```python
# loop using enumerate()
for i,v in enumerate(days):
    print(i,v)
```

```python
# use list comprehension
[print(x) for x in days]
```

### Cloning or copying a tuple

```python
t = (1,2,3)
```

```python
# assigment statement
a = t
print(a)
```

```python
# use slicing
b = t[:]
print(b)
```

```python
# use 'tuple()' function
c = tuple(t)
print(c)
```

```python
# use 'copy()' method
d = t.copy()
print(d)
```
