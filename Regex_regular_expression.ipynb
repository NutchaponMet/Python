## กรณีต้องการให้ตามด้วยเครื่องหมายวรรคตอน

```python
df[df['Title'].str.contains('man\b', case=False)]
```

```python
# \b เป็นอักษรพิเศษหากต้องการให้ใช้ได้จะต้อง \ ใส่เครื่องหมาย escape เพิ่มเข้าไปอีกตัวหนึ่งเพื่อแสดงให้เห็นว่าเป็นช่องว่าง
df[df['Title'].str.contains('man\\b', case=False)]
```

```python
# หากไม่ต้องการใส่ \\ เราก็จะต้องใส่ r' ' ให้ python อ่านข้อความเป็นแบบ raw string
df[df['Title'].str.contains(r'man\b', case=False)]  # กรณีนี้จะหมายถึงหาคำว่า man และตามด้วยช่องว่าง
# \s หมายถึงช่องว่างเท่ากัน , \b หมายช่องว่างและเครื่องหมายอะไรก็ได้ เช่น [- , ' _ :]
```

```python
# กรณีนี้จะหมายถึงช่องว่าง ' ' และตามด้วยคำว่า man$

df[df['Title'].str.contains(r'\bman$', case=False)]
```

```python
# กรณีขึ้นต้นด้วย man ตามด้วยอะไรก็ได้

df[df['Title'].str.contains(r'^man', case=False)]
```

```python
# กรณีขึ้นต้นด้วย man ตามด้วยช่องว่าง

df[df['Title'].str.contains(r'^man\b', case=False)]
```

```python
# กรณีหน้าคและหลังคำว่า man เป็นและช่องว่าง

df[df['Title'].str.contains(r'\bman\b', case=False)]
```

## กรณีใช้เครื่องหมาย [ ]

```python
df[df['Title'].str.contains(r'[1234man]', case=False)] # หมายถึงข้อมูลที่มีตัวอักษรในอยู่ประโยชน์ตามที่กำหนดไว้ใน [ ]
df[df['Title'].str.contains(r'[1-4man]', case=False)] # เขียนแบบที่ 2
```

```python
# การนำมาปรับใช้ กรณีเลือกตัวใดตัวหนึ่ง

df[df['Title'].str.contains(r'wom[ae]n', case=False)]
```

## กรณีใช้ OR หรือเครื่องหมาย |

```python
df[df['Title'].str.contains(r'bat|spider|ant|iron', case=False)]
```

```python
df[df['Title'].str.contains(r'(bat|spider|ant|iron)man', case=False)]
```

เครื่องหมาย * เมื่อใส่ตามหลังจะหมายถึงจะมีหรือไม่ก็ได้

```python
df[df['Title'].str.contains(r'(bat|spider|ant|iron)\s*man', case=False)]
# กรณีนี้ \s* หมายถึงช่องวางจะมีหรือไม่มีก็ได้
```

อย่างก็ตามการเขียนแบบ **(bat|spider|ant|iron)** จะต้องเปลี่ยนไปใช้ `str.extract.` แทนในอนาคต

```python
df[df['Title'].str.contains(r'(bat|spider|ant|iron)[\-\s]*man', case=False)]
```

```python
df[df['Title'].str.contains(r'^(bat|spider|ant|iron)', case=False)]
```

## **\d** หมายถึง ตัวเลข (digit)

```python
df[df['Title'].str.contains(r'\d', case=False)] # หมายถึงประโยคที่มีตัวเลขอยู่ข้างใน
# หรืออีกความหมายหนึ่งของ \d จะหมายถึง [0-9]
```

```python
# กรณีต้องการลงท้ายด้วยตัวเลข 2 ตัวเลขอะไรก็ได้ใช้เครื่องหมาย {} ตามหลัง \d

df[df['Title'].str.contains(r'\d{2}$', case=False)]
```

```python
df[df['Title'].str.contains(r'^\d{2}', case=False)] # กรณีขึ้นต้นด้วยตัวเลข 2 ตัวเลขอะไรก็ได้
```

```python
df[df['Title'].str.contains(r'(3|III|three)', case=False)]
```

```python
# กรณีต้องการแค่คำที่ระบุไว้ในวงเล็บเท่านั้น 3 อย่างเดียว

df[df['Title'].str.contains(r'\b(3|III|three)', case=False)]
```

```python
# กรณี fillter 2 columns

df[(df['Title'].str.contains('iron|avenger', case=False)) & (df['Actors'].str.contains('chris',case=False))]
```

### กรณี `AND` กับเครื่องหมาย lookahead `?=`
กรณีต้องการหาหนังแสดงที่ชื่อ Downey และ Evans

```python
# . หมายถึง ตัวอักษรอะไรก็ได้ ส่วน * หมายถึง มีตั้งแต่ 0 ตัวขึ้นไป
df[df['Actors'].str.contains('.*Downey.*Evans', case=False)]
```

```python
df[df['Actors'].str.contains('.*Evans.*Downey', case=False)]
```

```python
df[df['Actors'].str.contains('(.*Evans.*Downey)|(.*Downey.*Evans)', case=False)]
```

```python
# การนำเครื่องหมาย ?= (lookahead) มาปรับใช้
df[df['Actors'].str.contains('(?=.*Downey)(?=.*Evans)', case=False)]

# กรณีจะหมายถึง Downey จะมาก่อน Evans หรือ Evans มาก่อน Downey ก็ได้
```

Quantifiers

```python
# แต่ถ้าอยากให้มีแต่ i จำนวน 2 ตัวเท่านั้นจะต้องเขียนแบบนี้

df[df['Title'].str.contains(r'\bi{2}\b',case=False)]
```

# การใช้ str.extract, str.split กับการทำ clean and transform data

### pandas: `str.extract, str.split`

- use regular expression to clean and transform data

```python
import pandas as pd
print(f'pandas version: {pd.__version__}')
print(f'time: {pd.Timestamp.now()}')
```

```python
url='https://github.com/prasertcbs/basic-dataset/raw/master/regexp_extract_split.csv'
df=pd.read_csv(url)
df
```

## extract movie title and year

```python
# ส่วนแรก '(.{1,})' หมายถึงข้อความในส่วนแรก
# ส่วนสอง '\((\d{4})\)' หมายถึงข้อความที่มีวงลเล็บครอบอยู่

df.title_year.str.extract('(.{1,})\s\((\d{4})\)')
df.title_year.str.extract('(.+)\s\((\d{4})\)') # วิธีเขียนที่สอง
```

```python
dm=df.title_year.str.extract('(.+)\s\((\d{4})\)') # วิธีเขียนที่สอง
dm.columns=['title','year']
dm['year'] = dm['year'].astype(int)
dm
```

```python
dm.info()
```

## extract movie length

```python
df
```

```python
dt = df['length'].str.extract('(\d+)\shour\(s\)\s(\d+)\sminute\(s\)').astype(float)
```

```python
dt.columns = ['hour','minute']
dt
```

```python
dt['total_min'] = dt['hour'] * 60 + dt['minute']
dt
```

## extract budget and revenue ##

```python
df
```

```python
df.budget_revenue.str.replace(',','').str.split('/', expand=True).astype(float)
```

```python
ds = df.budget_revenue.str.replace(',','').str.split('/', expand=True).astype(float)
ds.columns = ['budget','revenue']
ds['profit'] = ds['revenue'] - ds['budget']
ds
```

## กรณีใช้ string extract

```python
dtemp = df.budget_revenue.str.replace(',','').str.extract('(.+)/(.+)').astype(float)
dtemp.columns = ['budget','revenue']
dtemp['profit'] = dtemp.revenue - dtemp.budget
dtemp
```

```python
df = pd.concat([df,dm,dt,dtemp], axis=1)
df
```

# การใช้ Regular Expression กรองคำที่เป็นภาษาไทย

```python
df = pd.read_csv('https://github.com/prasertcbs/basic-dataset/raw/master/gem.txt', names=['text'])
df
```

```python
from IPython.display import IFrame
IFrame('https://jrgraphix.net/r/Unicode/0E00-0E7F', width=800, height=400)
```

```python
de = df['text'].str.extract('(\w+)')
de.columns = ['eng_name']
```

```python
dt= df['text'].str.extract(r'([\u0E00-\u0E7F]+)')
dt.columns = ['subthai']
```

```python
dm= df['text'].str.extract(r'([\u0E00-\u0E7F]+$)')
dm.columns = ['meaning']
```

```python
df = pd.concat([df,de,dt,dm], axis=1)
df
```
