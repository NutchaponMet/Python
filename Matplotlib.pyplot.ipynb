### Working with Grid

```python
# Create figure
fig = plt.figure(figsize=(10,8))

# Convert to grid
ax1 = plt.subplot2grid((2,2), (0,0))
ax2 = plt.subplot2grid((2,2), (0,1))
ax3 = plt.subplot2grid((2,2), (1,0), colspan=2)

# Draw chart on ax1
x1 = np.arange(0,10,0.5)
y1 = 3*x1 +5
ax1.plot(x1,y1)

# Draw chart on ax2
x2 = np.arange(0,10,0.5)
y2 = 8*(x2**2) - 2*x2 - 10
ax2.plot(x2,y2, color='deepskyblue')

# Draw chart on ax3
x3 = np.arange(-10,10,0.5)
y3 = -(x3**3) + (x3**2) + x3 - 10
ax3.plot(x3, y3, color='black')
```

###  Chart  axes 

```python
from numpy.random import default_rng

# Create figure
fig = plt.figure(figsize=(10,8))

# Convert to grid
ax1 = plt.subplot2grid((2,2), (0,0))
ax2 = plt.subplot2grid((2,2), (0,1))
ax3 = plt.subplot2grid((2,2), (1,0), colspan=2)

# Draw chart on ax1
x1 = np.arange(0,10,0.5)
y1 = 3*x1 +5
ax1.plot(x1,y1)

# Draw chart on ax2
x2 = np.arange(0,10,0.5)
y2 = 8*(x2**2) - 2*x2 - 10
ax2.plot(x2,y2, color='deepskyblue')

# Draw chart on ax3
x3 = np.arange(-10,10,0.5)
y3 = -(x3**3) + (x3**2) + x3 - 10
ax3.plot(x3, y3, color='black')

rng = default_rng()
y4 = y3 + rng.integers(low=-200, high=200, endpoint=True, size=40)
ax3.scatter(x3, y4, color='red')
```

```python
# Visualization

# Prepare Data
displ = df.displ
cty = df.cty
hwy = df.hwy

# Create figure
fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2,figsize=(12,4), sharey=True, sharex=True)

# Convert to grid
# ax1 = plt.subplot2grid((1,2), (0,0))
# ax2 = plt.subplot2grid((1,2), (0,1))
# ax3 = plt.subplot2grid((2,2), (1,0), colspan=2)


#figure 1
# Create scatter plot
scatter_cty = ax1.scatter(cty, displ, s=10, c=displ , cmap='cool',)

# Set figure
ax1.set_xlabel('cty')
ax1.set_ylabel('displ')
ax1.set_title('displ vs cty')
plt.colorbar(scatter_cty, label='displacement',
              orientation='vertical', location='right', pad=.01)

#figure 2
# Create scatter plot
scatter_hwy = ax2.scatter(hwy, displ, s=10, c=displ, cmap='cool')

# Set figure
ax2.set_xlabel('hwy')
# ax2.set_ylabel('displ')
ax2.set_title('displ vs hwy')
plt.colorbar(scatter_hwy, label='displacement',
              orientation='vertical', location='right', pad=.01);
```
