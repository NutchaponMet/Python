# Lambda Function 

### 2.1 Lambda function as function parameter 

#### 2.1.1 Use with `map()` function

```python
rects = [(1,2),(3,4),(5,6)]
list(map(lambda x: x[0]*x[1], rects))
```

- Can we use the normal defined function instead?

answer is Yes!!

```python
def get_rect_area(x):
    return x[0]*x[1]

list(map(get_rect_area, rects))
```

#### 2.1.2 Use with `reduce()` function

```python
from functools import reduce

reduce(lambda x,y: x + y, [1,2,3,4,5,6,7,8,9])
```

#### 2.1.3 Use with `filter()` function

```python
name = ['Dave', 'Nick', 'Tom', 'Mike']

list(filter(lambda s: s[0] in ['D','T'], name))
```

### 3. List Comprehension as Alternative to Lambda Function
--------------------------------------------------------------------

```python
rects = [(1,2),(3,4),(5,6)]


# Lambda
print(list(map(lambda x: x[0]*x[1], rects)))

# list Comprehension

print([x[0]*x[1] for x in rects])
```

```python
seq = [1,2,3,4,5,6,7,8,9]

# Lambda
print(reduce(lambda x,y: x + y, [1,2,3,4,5,6,7,8,9]))

# list

print(sum(seq))
```

```python
name = ['Dave', 'Nick', 'Tom', 'Mike']

# Lambda
print(list(filter(lambda s: s[0] in ['D','T'], name)))

# List
print([x for x in name if x[0] in ['M','N']])
```
