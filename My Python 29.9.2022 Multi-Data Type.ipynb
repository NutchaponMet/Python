## Working with Different Data types (1)

Let's observe the **data types** of expression involing operands of the **same data types**

```python
print(type(78 + 22))
```

```python
print(type(1.5 - 1.6))
```

```python
print(type(3+5j + 1+2j))
```

We can see that binary Arithmetic operators with *both operands being of similar data types* result in values with the *same data types of the operands*.

Except for the *division operator*.

```python
print(type(2/2))
```

```python
print(type(2.0/2.0))
```

We can see that whether the operands were either `int` or `float`, the results were always `float`. this makes sense since the result of a division can have *decimal point*.

Now, let's try dividing complex numbers.

```python
c1 = 4+4j
c2 = 2+2j
print(c1/c2)
print(type(c1/c2))
```

This, again, made sense since a division of two complex numbers can result in another complex numbers.

# Polymorphic operators

### Arithmetic Addition & String Concatenation

the `+` symbol performs *Arithmetic Addition* when both operands are *numbers*.

```python
print(2+2)
print(2.2+2.5+3.4)
```

The `+` symbol performs *String Concatenation* when both operands are `str`.

```python
s = "1" + "2" + "3"
print(s)
print(type(s))
```

Mixing `str` and *numbers* as operands of `+` results in an **error** since the operator *dose not support* such a data type combination.

### Arithmetic Multiplication & String Multiplication

The `*` symbol performs *Arithmetic Multiplication* when both operands are *numbers*.

```python
print(2*-0.5)
print(110*110*110)

print(type(2*-0.5))
print(type(110*110*110))
```

The `*` symbol performs *String Multiplication* when one operand is `str` while the other is `int`.

```python
print("♥"*10)
print(10*"♫")
```

In *String Multiplication*, the number **must** be `int` only.

# Working with Different Data Types (3)

## Implicit Data Type Conversion
- Occurs *automatically*
- **Implicit Data Type Conversion** always promote data type capable of holding *less information* to data type capable of holding *more information*.

```python
print(type(32 * 5.0))
```

In the above statement, the `int` value of 32 was **converted** to `float` and then **multiplied** with 5.0.

```python
print(type(5 * (3+2j)))
print(type((1+1j) + -0.5))
```

In the code above, both `int` and `float` were **converted** to `complex` before being used as operands in their respective statement.

## Explicit Data Type Conversion
We use *typecasting function* to converted data to the desired data types.

Typecasting function inculde (but not limited to):
- `int()`
- `float()`
- `complex()`
- `str()`
- `bool()`

**An example situation in which *explicit data type conversion* is needed.**

```python
# This program calculates n^10, where n is a number entered by the user.
'''
# Case 1
n = input("Please enter n:")
result = float(n) ** 10 #<-- Now result's data type is [float]
# Show the output as n = [the input] --> n^10 = [the result]
print(result)
#-------------------------------------------------------------
'''
# Case 2
n = input("Please enter n:")
result = float(n) ** 10 #<-- Now result's data type is [float]
# Show the output as n = [the input] --> n^10 = [the result]
print("n = " + n + " --> n^10 = " + str(result))
```

## `int()`

```python
print(int(1.8)) # <-- float แปลงเป็น int ได้
# print(int(2+2j)) <-- Complex แปลงข้าม float เป็น int ไม่ได้ 
print(int("12")) # <-- str แปลงเป็น int ได้เนื่องจากหน้าตาคล้าย int
#print(int("12.5")) <-- str แบบ float แปลงเป็น int ไม่ได้เนื่องจากหน้าตาไม่คล้ายกัน
#print(int("Hello")) <-- ตายตัว str แปลงเป็น int ไม่ได้
print(int(True))
print(int(False))
```

# `float()`

```python
print(float(1)) 
# print(float(2+2j)) <-- Complex ไม่สามารถแปลงลงมาเป็น float ได้
print(float("12"))
print(float("12.5"))
print(float(True))
print(float(False))
```

# `complex()`

```python
print(complex(1))
print(complex(1.8))
print(complex("12"))
print(complex("12.5"))
print(complex(True))
print(complex(False))
```

To create a new complex numbers with `complex()`, we can use `complex(re,im)` (called a *constructor function*).

```python
print(complex(1,1))
print(complex(1.5,-1.0))
print(complex(1,0))
print(complex(0,1))
```
