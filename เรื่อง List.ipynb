# LIST
Lists are another data type in python to store multiple **items** in a single variable.

## Creating Lists
- Lists can be created using square brackets `[ ]`.

```python
colors = ["red", "green", "blue"]
print(colors)
```

```python
type(colors)
```

### List items are "ordered", addressing by an *index*.
- Python uses *zero-based* indexing. The first item has an index `0`,the secound item has index `1`,and so on.

- To access an item in the list using **index**; we use a square bracket: `[ ]`

```python
print(colors[0])
print(colors[2])
```

In Python, the index can be negative; we can use index from `-1` and below to address from tail to head of the list.




```python
print(colors[-1])
```

- The error `list index out of range` arises if you access invalid indices in your list.

```python
print(colors[10])
```

### Slicing
- We can also use **slicing** to access a *range of items* by defining the `start`,`stop`,and `step` parameters.

```python
# SYNTAX: list[start:stop:step]
# "start" refers to the "start" position of slice
# "stop" refers to the "end" position of the slice
# "step" refers to the "increment"

letters = 'abcdefghijklmnopqrstuvwxyz'
l = list(letters)
l_slice = l[2:5:1]
s_slice = letters[2:5:1]

print(l_slice)
print(s_slice)
print(l)
```

```python
# default "step" = 1
print(l[2:5])
```

```python
# "step" allows us to take each nth-element within a start:stop range.
print(l[2:10:2])
```

```python
# defaul "start" = 0
print(l[:5])
```

```python
# defaul "stop" = end of the line
print(l[2:])
```

```python
# "start" & "stop" can be negetive index
print(l[20:-1])
```

```python
print(l[-5:])
```

```python
# sample of every other items
print(l[::2])
```

```python
# to reverse the order of the list ใช้ได้ทั้งเป็น string และ list
print(l[::-1])
print(letters[::-1])
```

### Lists are "mutable"; its items can be modified.
### การเปลี่ยนแปลง = Add, Update, Delete
- We can add, update, delete items in a list after it has been created.
- To **add** more item(s) to an existing list:
    - `append()` adds an item to the end of the list.
    - `extend()` adds multiple items to the end of the list.
    - `insert()` insert an item at specified index.
    - `+` combines lists.
   

```python
#Add an item using append()
colors.append('yellow')
print(colors)
```

```python
colors.append('123')
print(colors)
```

```python
# Add an iterable object to the list using extend()
colors_mixed = ['yellows','cyan','magenta']

colors.extend(colors_mixed)
print(colors)
```

```python
colors.extend('xyz')
print(colors)
```

```python
# Insert an item at specified index using insert() เป็นการแทรกค่าใหม่ไปแทนที่ค่าเดิม
colors.insert(1,'pink')
print(colors)
```

```python
# Combine lists using an operator '+'
# Operator '+' จะไม่กระทบกับ list ตัวตั้งต้น

major_colors = ['red','green','blue']
mix_colors = ['yellow','cyan','magenta']

colors = major_colors + mix_colors
print(colors)
```

- To **update** item(s) in a list

```python
# Update an item in a list
colors[1] = 'gray'
print(colors)
```

```python
# Update multiple items within a specific range
colors[1:3] = ['brown', 'orange']
print(colors)
```

```python
# กรณีที่ Update ค่าเกิน ตัวที่เกินมาจะเป็นในรูปแบบของการแทรกค่าเข้าไป
colors[1:3] = ['maroon','lime','purple']
print(colors)
```

```python
# กรณีนี้จะเป็นการนำค่า 'bronze' ไปแทนค่าที่อยู่ใน list ตำแหน่งที่ 1-2
colors[1:3] = ['bronze']
print(colors)
```

```python
# Insert at the head กรณีแทรกไว้ที่หัวแถว
colors[:0] = ['gold','silver']
print(colors)
```

```python
# Insert at the tail กรณีต่อท้ายแถว
colors[len(colors):] = ['navy','white']
print(colors)
```

- To **delete** item(s) from a list.
    - `remove()` removes an list by value.
    - `pop()` removes an item by index and get its value.
    - `del` removes items by index or  slice.
    - `clear()` removes all items.
    

```python
# Remove the specified item by value using 'remove()'
colors.remove('navy')
print(colors)
```

```python
# Remove the specified item by index using 'pop()'
colors.pop(5)
print(colors)

# Remove the last item:
colors.pop()
print(colors)
```

```python
# Get pop() return value
c = colors.pop()
print(c)
print(colors)
```

```python
# Delete the specified item by index using 'del' function
del colors[0]
print(colors)
```

```python
# Delete the specified items using slicing
del colors[:2]
print(colors)
```

```python
# Delete completely
del colors
print(colors)
```

```python
# Empty the list using 'clear()'
colors = ['red','green','blue','black']
print(colors)
```

```python
colors.clear()
print(colors)
```

### Lists are "iterable" objects.
- We can **iterate** through lists.

```python
planets = ['Mercury','Venus','Earth','Mars','Jupiter','Saturn','Uranus','Neptune']
```

```python
# use for loop
for i in planets:
    print(i)
```

```python
# use while loop
i = 0;
while i < len(planets):
    print(planets[i])
    i += 1
```

```python
# loop through the index number
for i in range(len(planets)):
    print(planets[i])
```

```python
# loop using enumerate()*** สำคัญ

for i,v in enumerate(planets):
    print(i,v)
```

```python
# use list comprehension
[print(x) for x in planets]
```

### Finding item(s) in a list.

```python
friends = ['cad','be','pui','am','tua','ead','pui']
```

```python
# index สามารถรับ parameter ได้หลายลักษณะ
# Find the position of a given item in a list using 'index()' method.
friends.index('pui') #return the index of the first occurrence
```

```python
friends.index('pui',3) # starting index is given
```

```python
friends.index('pui',2,5)
```

```python
# Check of an item exists in a list
if 'pui' in friends:
    print('Yes, she is in the position#', friends.index('pui'), 'of the list.')
```

```python
# Count the number of occurence for a given item in a list using 'count()' method.
friends.count('pui')
```

```python
friends.count('pinky')
```
