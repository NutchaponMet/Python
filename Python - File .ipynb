```python
f = open('Mylist.txt')
```

```python
print(f)
```

### Access Modes
- `r`: read only (default)
- `w`: write only
- `r+`: read  & write
- `w+`: write & read
- `a`: append **()**
- `t`: text mode(default)
- `b`: binary mode

### Read & write files

#### Read a flie
- `read(<size>)`: read *size* characters or bytes from the file. if no argument, it will read entire file.
- `readline()`: reads a single line from the file, with `\n` at the end (except the last line)
- `readlines()`: reads all lines and return a list of each line.
- Loop to read each line:
   
    `for line in file odject:`
            ...

```python
data = f.read()
print(data)
```

```python
data = f.read(5)
print(data)
```

```python
data = f.readline()
print(data)
```

```python
print(len(data))
```

```python
data = f.readlines()
print(data)
```

```python
for ln in f:
    print(ln)
```

```python
for ln in f:
    print(ln, end='')
```

### Write to a file
- `write(<string>)` : write the *string* to the file and return a number of written characters.
- `writelines()` : write a list of lines to the file.

```python
f.write('Bye')
```

```python
f = open('Mylist.txt', 'a')
```

```python
f.write('Bye')
```

```python
f.close()
```
