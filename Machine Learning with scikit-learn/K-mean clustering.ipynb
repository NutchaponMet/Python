# K-mean clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
df = pd.read_csv(url, names=['sepal_length','sepal_width','petal_length','petal_width','species'])
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df.describe().T
```

# Scikit-learn: KMeans Clustering

```python
sns.pairplot(df, vars=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
            hue='species',
            plot_kws={'alpha':.4});
```

```python
sns.scatterplot(data=df, x='petal_length', y='petal_width',
               hue='species')
```

```python
rx=np.random.uniform(1,7,3)
rx
```

```python
ry=np.random.uniform(0,2.5,3)
ry
```

```python
sns.scatterplot(data=df, x='petal_length', y='petal_width',
               hue='species')
plt.scatter(rx, ry, color='.1', marker='D')
```

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=3)
model
```

```python
X=df[['petal_length','petal_width']]
model.fit(X)
```

```python
model.cluster_centers_
```

```python
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1])
```

```python
sns.scatterplot(data=df, x='petal_length', y='petal_width',
               hue='species',
               alpha=.5)
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], color='.1',marker='D')
```

```python
model.labels_
```

```python
df['species'].values
```

```python
df['cluster'] = model.labels_
df
```

```python
pd.crosstab(df['species'], df['cluster'])
```

```python
sns.scatterplot(data=df, x='petal_length', y='petal_width',
               hue='species',
               alpha=.5)
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], color='.1',marker='D')
```

```python
sns.scatterplot(data=df, x='petal_length', y='petal_width',
               hue='cluster',
               alpha=.5,
               palette=['green','blue','orange'])
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], color='.1',marker='D')
```

```python
fig, ax= plt.subplots(1, 2, figsize=(12,5), sharey=True, sharex=True)

g1=sns.scatterplot(data=df, x='petal_length', y='petal_width',
               hue='species',
               alpha=.5, ax=ax[0])
g1.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], color='.1',marker='x')

g2=sns.scatterplot(data=df, x='petal_length', y='petal_width',
               hue='cluster',
               alpha=.5,
               palette=['green','blue','orange'], ax=ax[1])
g2.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], color='.1',marker='D')
```

```python
# predict
model.predict([
    [2, 3],
    [4, 6],
    [0,6.2]
])
```
