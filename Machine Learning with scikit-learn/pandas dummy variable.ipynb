```python
import pandas as pd
```

```python
df=pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/msleep.csv')
df
```

```python
df.info()
```

### check NA in *vore* columns

```python
df[df.vore.isna()]
```

```python
df.dropna(subset=['vore'], axis='index', inplace=True)
df
```

### create dummy variables

```python
dummies = pd.get_dummies(df['vore'])
dummies
```

```python
dummies.columns
```

```python
dm = pd.concat([df, dummies], axis='columns')
dm
```

## sklearn: LinearRegression 

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
dt=dm[['vore','bodywt','carni', 'herbi', 'insecti', 'omni', 'sleep_total']].dropna(axis='index')
X = dt[['bodywt', 'herbi', 'insecti', 'omni']]
```

```python
model.fit(X, df.sleep_total)
```

```python
model.score(X, df.sleep_total) # R-squared
```

```python
model.coef_
```

```python
X
```

## Statmodels: LinearRegression 

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
model_a = smf.ols(formula='sleep_total ~ bodywt + insecti + herbi + omni',
                 data=dt).fit()
```

```python
print(model_a.summary())
```

```python
model_a.pvalues
```

```python
model.intercept_
```

```python
model.coef_
```

## use `C()` in Statmodels to create dummy variables internally

```python
model_b=smf.ols(formula='sleep_total ~ bodywt + C(vore)', data=dt).fit()
```
