```python
m = input("คำ 3 คำ: ").strip()
m1, m2, m3 = m.split()
print(m1)
print(m2)
print(m3)
print(type(m1))
print(type(m2))
print(type(m3))
print(m1 + " " + m2 + " " + m3 + ".")
print(len(m1))
print(m2.lower())
print(m3.upper())
print("Nutchapon" in m1)
print("yosita" in m2)
print("Print" in m3)
M1 = m1.upper()
M2 = m2.upper()
M3 = m3.upper()
print(M1, M2, M3, sep=" ")
```

```python
M1, M2, M3 = M1.lower(), M2.lower(), M3.lower()
print(M1,M2,M3, sep=" ")
M = M1 + " " + M2 + " " + M3 + "."
print(M)
print(len(M))
```
