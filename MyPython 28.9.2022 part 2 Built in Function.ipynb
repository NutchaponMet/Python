## Built-in Function มากับ Python อยู่แล้ว เรียกใช้ได้เลย

### Built-in Function  Examples

#### ads()

```python
print(abs(5))
print(abs(-5))
```

```python
print(abs(5.0))
print(abs(-5.5))
```

```python
print(abs(-1+2j))
print(abs(3+4j))
```

```python
print(abs("Nut"))
```

### bin()

```python
x = bin(200)
print(x)
print(type(x))
```

```python
x = bin(200.2)
```

### input([prompt])

```python
name = input()
```

```python
name = input("Enter your name: ")
print(name)
```

### pow(base, exp)
ฟังก์ชัน ยกกำลัง

```python
print(pow(2,5))
print(pow(5,2))

print(pow(exp=6,base=7))
```

## function max
max(arg1, arg2, *args [, key])  *agrs หมายถึง argument กี่ตัวก็ได้ หรือไม่ใส่ก็ได้ ใส่ได้ทั้ง `int`, `float`, `str`,

```python
print(max(8,9.5))
```

```python
print(max(10,9,-3))
```

```python
print(max(2,3,5,-4,1.0))
```

```python
print(max("A","B","C","D","E","F"))
```
